```python
import pandas as pd
import numpy as np
from datetime import date, datetime, timedelta
from sklearn import linear_model
import os
from scipy import stats

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
CBE = pd.read_table("cbe.dat")
```

```python
CBE.head()
```

```python
#add date/time
def perdelta(start, end, delta):
    curr = start
    while curr < end:
        yield curr
        curr += delta
fulldate = []
for result in perdelta(date(1958, 1, 1), date(1991, 1, 1), timedelta(days=1)): 
    fulldate.append(result)
```

```python
len(fulldate)
```

```python
fulldate[-1]
```

```python
fdate = [d for d in fulldate if d.day == 1]
```

```python
fdate[:5]
```

```python
#append list of dates to data
len(fdate)
```

```python
len(CBE)
```

```python
CBE["fdate"] = fdate
```

```python
CBE.head()
```

```python
CBE.index = CBE.fdate
```

```python
CBE.head()
```

```python
fig = plt.figure(figsize = (20,10))
# (###) means row, column, plot position
ax1 = fig.add_subplot(311)
ax2 = fig.add_subplot(312)
ax3 = fig.add_subplot(313)

## the data
x1 = CBE['fdate']
y1 = CBE['choc']

x2 = CBE['fdate']
y2 = CBE['beer']

x3 = CBE['fdate']
y3 = CBE['elec']

ax1.plot(x1,y1,color='blue')
ax1.tick_params(labelsize=15)
ax1.set_title("Chocolate", fontsize=20)

ax2.plot(x2,y2,color='blue')
ax2.tick_params(labelsize=15)
ax2.set_title("Beer", fontsize=20)

ax3.plot(x3,y3,color='blue')
ax3.tick_params(labelsize=15)
ax3.set_title("Electricity", fontsize=20)

plt.show()
```
